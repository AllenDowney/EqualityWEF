```python
from utils import make_weight_table

tables = [
    make_weight_table(primary, 'primary'),
    make_weight_table(secondary, 'secondary'),
    make_weight_table(tertiary, 'tertiary'),
]
weight_table = pd.concat(tables)
weight_table
```

```python
def compute_weights(table):
    inv_std = table["inv std"]
    return table.assign(weight=inv_std / inv_std.sum())

keys = ["original", "revised"]
pd.concat([compute_weights(weight_table[key]) for key in keys], axis=1, keys=keys) 
```
